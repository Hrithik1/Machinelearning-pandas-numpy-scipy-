# Standard Deviation & Variance

```python
import numpy as np
import matplotlib.pyplot as plt

income=np.random.normal(40,50,100)

plt.hist(income,50)
plt.show()
```

```python
income.std()
```

```python
income.var()
```

# Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.random.uniform(-20,20,100000)

plt.hist(a,50)
plt.show()
```

# Normal/Gaussian Distribution

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(-3,3,0.005)
plt.plot(a,norm.pdf(a))
```
